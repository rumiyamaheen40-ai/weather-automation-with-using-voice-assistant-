7) Swap Highest & Lowest Temperature

```python
temps = [28, 32, 25, 30, 35, 27, 29]

highest = max(temps)
lowest = min(temps)

hi = temps.index(highest)
lo = temps.index(lowest)

temps[hi], temps[lo] = temps[lo], temps[hi]

print("Updated temperatures:", temps)
```

8) Mean of Numbers

```python
numbers = [10, 20, 30, 40, 50]
mean = sum(numbers) / len(numbers)
print("Mean:", mean)
```

9) Median of Numbers

```python
numbers = [10, 20, 30, 40, 50]
numbers.sort()

n = len(numbers)
if n % 2 != 0:
    median = numbers[n//2]
else:
    median = (numbers[n//2 - 1] + numbers[n//2]) / 2

print("Median:", median)
```

10) Dictionary Program

```python
data = {"name": "Alex", "age": 21}
key = "employee_id"

if key in data:
    data.pop(key)
else:
    data[key] = 4503233

print("Updated dictionary:", data)
```
